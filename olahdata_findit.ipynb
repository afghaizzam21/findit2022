```python
df_train = pd.read_csv('train.csv',
                   sep = '|',
                   engine = 'python')
```

```python
df_train.dropna(subset=["salary"],inplace=True)
df_train.reset_index(drop = True,inplace=True)
```

```python
df_train.salary = df_train.salary.mask(df_train.salary_currency == "USD", df_train.salary*15000)
df_train.salary_currency =df_train.salary_currency.mask(df_train.salary_currency == "USD", "IDR")
```

```python
import warnings
```

```python

warnings.filterwarnings("ignore", 'This pattern has match groups')
conditions = [
    df_train["education_level"].str.contains("(S3)"),
    df_train["education_level"].str.contains("(S2)",) & ~df_train["education_level"].str.contains("(S3)"),
    df_train["education_level"].str.contains("(S1)") & ~df_train["education_level"].str.contains("(S2)(S3)"),
    df_train["education_level"].str.contains("(Diploma)") & ~df_train["education_level"].str.contains("(S1)(S2)(S3)"),
    df_train["education_level"].str.contains("SMA, SMU/SMK/STM") & ~df_train["education_level"].str.contains("(Diploma)(S1)(S2)(S3)"),
    df_train["education_level"].str.contains("Tidak terspesifikasi")
]
values = ["S3","S2" ,"S1" ,"Diploma" ,"SMA" ,"Tidak terspesifikasi"]
df_train['pendidikan_terakhir'] = np.select(conditions, values)
```

```python
df_train["sertifikat_proffesional"] = np.where(df_train["education_level"].str.contains("Sertifikat Professional"), True, False)
```

```python
warnings.filterwarnings("ignore", 'This pattern has match groups')
conditionss = [
    df_train["company_size"].str.contains("1- 50 pekerja",na=False),
    df_train["company_size"].str.contains("51 - 200 pekerja",na=False),
    df_train["company_size"].str.contains("201 - 500 pekerja",na=False),
    df_train["company_size"].str.contains("501 - 1000 pekerja",na=False),
    df_train["company_size"].str.contains("1001 - 2000 pekerja",na=False),
    df_train["company_size"].str.contains("2001 - 5000 pekerja",na=False),
    df_train["company_size"].str.contains("Lebih dari 5000 pekerja",na=False),
]
valuess = [1,2,3,4,5,6,7]
df_train['ukuran_company'] = np.select(conditionss, valuess)
```

```python
df_train["bidang"] = df_train.job_function.str.split(",").str[0]
```

```python
df_train['total_huruf'] = df_train.job_description.str.len()
df_train['total_kata'] = df_train.job_description.str.split().str.len()
```
